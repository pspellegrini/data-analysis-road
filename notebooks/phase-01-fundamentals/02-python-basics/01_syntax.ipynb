## 1. Python executa linha por linha

Diferente de outras linguagens, Python lê e executa o código de cima para baixo, uma linha por vez.
Se uma linha tiver erro, ele para ali e avisa.

````python
# Python executa esta linha primeiro
print("Linha 1")

# Depois esta
print("Linha 2")

# E por último esta
print("Linha 3")
````

## 2. Indentação — a regra mais importante do Python

Em Python, **espaços no início da linha têm significado**. Eles definem blocos de código.

Analogia: imagine uma receita de bolo. Os passos "dentro" do forno ficam recuados:
```
Se o forno estiver quente:
    coloque o bolo
    aguarde 40 minutos
Retire o bolo
```

Em Python usa-se **4 espaços** (ou 1 Tab) por nível de indentação.

````python
# Exemplo correto de indentação
temperatura = 200

if temperatura > 180:
    print("Forno quente — pode colocar o bolo")  # 4 espaços de recuo
    print("Aguarde 40 minutos")                   # mesmo nível

print("Retire o bolo")  # fora do bloco if
````

## 3. Comentários

Comentários são notas que você deixa no código. Python ignora tudo depois do `#`.
Use para explicar **por que** algo é feito, não **o que** é feito (o código já mostra o quê).

````python
# Isso é um comentário — Python ignora esta linha

preco = 100  # preço em reais
desconto = 0.10  # 10% de desconto

# Calcula o valor final após desconto
preco_final = preco * (1 - desconto)

print(preco_final)
````

## 4. `print()` — exibindo informações

`print()` é a função que mostra algo na tela. É sua principal ferramenta para entender o que está acontecendo no código.

````python
# print com texto simples
print("Olá, mundo!")

# print com número
print(42)

# print com variável
nome = "Paulo"
print(nome)

# print com texto + variável (f-string — o jeito moderno)
idade = 30
print(f"Meu nome é {nome} e tenho {idade} anos")

# print com separador personalizado
print("Python", "é", "incrível", sep="-")

# print sem pular linha ao final
print("Linha sem quebra", end=" ")
print("continua aqui")
````

## 5. `input()` — recebendo dados do usuário

`input()` pausa o programa e espera o usuário digitar algo.
**Importante:** tudo que vem do `input()` é texto (`str`), mesmo que o usuário digite um número.

````python
# Recebe o nome do usuário
nome = input("Qual é o seu nome? ")
print(f"Olá, {nome}!")
````

````python
# Recebe um número — precisa converter para int ou float
idade_texto = input("Qual é a sua idade? ")
idade = int(idade_texto)  # converte texto para número inteiro

print(f"Daqui a 10 anos você terá {idade + 10} anos")
````

## 6. Tipos de dados básicos

Python tem 4 tipos primitivos que você vai usar o tempo todo:

| Tipo | Nome | Exemplo |
|---|---|---|
| `int` | Inteiro | `42`, `-7`, `0` |
| `float` | Decimal | `3.14`, `-0.5`, `100.0` |
| `str` | Texto | `"Python"`, `'olá'` |
| `bool` | Verdadeiro/Falso | `True`, `False` |

Use `type()` para descobrir o tipo de qualquer valor.

````python
# Exemplos de cada tipo
numero_inteiro = 42
numero_decimal = 3.14
texto = "Python"
verdadeiro_ou_falso = True

# type() revela o tipo de cada variável
print(type(numero_inteiro))        # <class 'int'>
print(type(numero_decimal))        # <class 'float'>
print(type(texto))                 # <class 'str'>
print(type(verdadeiro_ou_falso))   # <class 'bool'>
````

## 7. Conversão entre tipos

Às vezes você precisa transformar um tipo em outro. Python oferece funções para isso:

- `int()` — converte para inteiro
- `float()` — converte para decimal
- `str()` — converte para texto
- `bool()` — converte para booleano

````python
# Texto para número
preco_texto = "29.90"
preco = float(preco_texto)  # agora é um número decimal
print(preco + 10)           # 39.9 — funciona como número

# Número para texto
ano = 2024
mensagem = "Ano: " + str(ano)  # precisa converter para concatenar
print(mensagem)

# Decimal para inteiro (perde a parte decimal — não arredonda)
altura = 1.78
print(int(altura))  # 1, não 2

# Valores que viram False quando convertidos para bool
print(bool(0))       # False — zero é falso
print(bool(""))      # False — texto vazio é falso
print(bool(None))    # False — nada é falso
print(bool(42))      # True — qualquer número diferente de zero é verdadeiro
print(bool("oi"))    # True — texto não vazio é verdadeiro
````
