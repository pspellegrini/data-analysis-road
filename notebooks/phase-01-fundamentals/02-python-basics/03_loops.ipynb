# Python Básico — Loops

## O que vamos aprender
Loops permitem **repetir** um bloco de código sem precisar escrever ele várias vezes.

Analogia: em vez de escrever "lave o prato" 10 vezes, você escreve "repita 10 vezes: lave o prato".

1. `for` — repetir um número determinado de vezes
2. `range()` — gerar sequências de números
3. `while` — repetir enquanto uma condição for verdadeira
4. `break` — sair do loop antes do fim
5. `continue` — pular para a próxima iteração
6. `enumerate()` — loop com índice
7. Loops aninhados

## 1. `for` — percorrendo uma sequência

O `for` percorre cada item de uma sequência (lista, texto, range...).

Sintaxe:
```python
for variavel in sequencia:
    # código que roda para cada item
```

````python
# Percorrendo uma lista de frutas
frutas = ["maçã", "banana", "laranja"]

for fruta in frutas:
    print(fruta)  # imprime cada fruta, uma por linha
````

````python
# Percorrendo letras de um texto
for letra in "Python":
    print(letra)
````

## 2. `range()` — gerando sequências numéricas

`range()` gera uma sequência de números sem precisar criar uma lista.

- `range(5)` → 0, 1, 2, 3, 4 (começa em 0, para antes do 5)
- `range(1, 6)` → 1, 2, 3, 4, 5 (começa em 1, para antes do 6)
- `range(0, 10, 2)` → 0, 2, 4, 6, 8 (passo de 2)

````python
# Contando de 0 a 4
for i in range(5):
    print(i)
````

````python
# Contando de 1 a 5
for i in range(1, 6):
    print(i)
````

````python
# Tabuada do 7
numero = 7

for i in range(1, 11):
    print(f"{numero} x {i} = {numero * i}")
````

## 3. `while` — repetindo enquanto a condição for verdadeira

Use `while` quando não sabe quantas repetições serão necessárias — o loop continua até a condição ser False.

Analogia: "**Enquanto** tiver sujeira no prato, continue lavando."

> ⚠️ **Cuidado com loop infinito**: se a condição nunca ficar False, o programa trava. Sempre garanta que algo dentro do loop vai mudar a condição.

````python
# Contagem regressiva
contador = 5

while contador > 0:
    print(contador)
    contador -= 1  # diminui 1 a cada volta — sem isso seria loop infinito!

print("Lançar!")
````

````python
# Validação de entrada — pede até o usuário digitar algo válido
senha = ""

while senha != "1234":
    senha = input("Digite a senha: ")

print("Acesso liberado!")
````

## 4. `break` — saindo do loop antes do fim

`break` interrompe o loop imediatamente, mesmo que ainda haja itens para percorrer.

````python
# Procurando um número em uma lista
numeros = [3, 7, 1, 9, 4, 6, 2]
alvo = 9

for numero in numeros:
    print(f"Verificando {numero}...")
    if numero == alvo:
        print(f"Encontrei o {alvo}!")
        break  # para de procurar — não precisa ver o resto
````

## 5. `continue` — pulando para a próxima iteração

`continue` pula o restante do código na iteração atual e vai direto para a próxima.

````python
# Imprimindo apenas os números pares de 1 a 10
for i in range(1, 11):
    if i % 2 != 0:  # se for ímpar...
        continue    # ...pula para o próximo
    print(i)        # só chega aqui se for par
````

## 6. `enumerate()` — loop com índice

Quando você precisa saber tanto o item quanto sua posição na lista, use `enumerate()`.

````python
# Listando produtos com numeração
produtos = ["notebook", "mouse", "teclado", "monitor"]

for indice, produto in enumerate(produtos):
    print(f"{indice + 1}. {produto}")  # +1 para começar em 1, não 0
````

## 7. Loops aninhados

Um loop dentro de outro. O loop interno completa todas as suas repetições para cada repetição do externo.

````python
# Tabela de multiplicação de 1 a 3
for i in range(1, 4):       # linhas: 1, 2, 3
    for j in range(1, 4):   # colunas: 1, 2, 3
        print(f"{i} x {j} = {i * j}")
    print("---")  # separador entre linhas
````
