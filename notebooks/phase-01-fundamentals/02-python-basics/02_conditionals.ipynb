# Python Básico — Condicionais

## O que vamos aprender
Condicionais permitem que o programa **tome decisões** — execute um trecho de código só quando uma condição for verdadeira.

É como uma bifurcação na estrada: dependendo da condição, o programa segue por um caminho ou outro.

1. `if` — se
2. `else` — senão
3. `elif` — senão se
4. Operadores de comparação
5. Operadores lógicos (`and`, `or`, `not`)
6. Condicionais aninhadas
7. Operador ternário (if em uma linha)

## 1. `if` — execute só se a condição for verdadeira

Sintaxe:
```python
if condição:
    # código que roda apenas se condição for True
```

Analogia: "**Se** estiver chovendo, leve guarda-chuva."

````python
# Verificando se a temperatura está alta
temperatura = 35

if temperatura > 30:
    print("Está quente! Beba água.")  # só roda se temperatura > 30

print("Fim do programa")  # sempre roda
````

## 2. `else` — o que fazer quando a condição é falsa

Sintaxe:
```python
if condição:
    # roda se True
else:
    # roda se False
```

Analogia: "**Se** estiver chovendo, leve guarda-chuva. **Senão**, use óculos de sol."

````python
# Verificando se o usuário é maior de idade
idade = 17

if idade >= 18:
    print("Acesso permitido")  # roda se idade >= 18
else:
    print("Acesso negado — menor de idade")  # roda se idade < 18
````

## 3. `elif` — múltiplas condições

Quando você tem mais de dois casos possíveis, use `elif` (abreviação de "else if").
Python testa cada condição em ordem e para na primeira que for True.

Sintaxe:
```python
if condição_1:
    ...
elif condição_2:
    ...
elif condição_3:
    ...
else:
    ...  # nenhuma das anteriores foi True
```

````python
# Classificação de nota escolar
nota = 75

if nota >= 90:
    conceito = "A"
elif nota >= 80:
    conceito = "B"
elif nota >= 70:
    conceito = "C"
elif nota >= 60:
    conceito = "D"
else:
    conceito = "F"

print(f"Nota {nota} → Conceito {conceito}")
````

## 4. Operadores de comparação

São usados dentro das condições para comparar valores:

| Operador | Significado | Exemplo | Resultado |
|---|---|---|---|
| `==` | igual a | `5 == 5` | `True` |
| `!=` | diferente de | `5 != 3` | `True` |
| `>` | maior que | `7 > 3` | `True` |
| `<` | menor que | `2 < 1` | `False` |
| `>=` | maior ou igual | `5 >= 5` | `True` |
| `<=` | menor ou igual | `4 <= 3` | `False` |

> **Atenção:** `=` é atribuição (dá um valor). `==` é comparação (verifica se são iguais).

````python
# Testando os operadores de comparação
x = 10

print(x == 10)   # True  — x é igual a 10?
print(x != 5)    # True  — x é diferente de 5?
print(x > 8)     # True  — x é maior que 8?
print(x < 8)     # False — x é menor que 8?
print(x >= 10)   # True  — x é maior ou igual a 10?
print(x <= 9)    # False — x é menor ou igual a 9?
````

## 5. Operadores lógicos: `and`, `or`, `not`

Permitem combinar múltiplas condições:

| Operador | Significado | Verdadeiro quando... |
|---|---|---|
| `and` | e | **ambas** as condições são True |
| `or` | ou | **pelo menos uma** condição é True |
| `not` | não | inverte — True vira False e vice-versa |

````python
# and — as duas condições precisam ser verdadeiras
idade = 25
tem_carteira = True

if idade >= 18 and tem_carteira:
    print("Pode dirigir")
else:
    print("Não pode dirigir")
````

````python
# or — basta uma das condições ser verdadeira
dia = "sábado"

if dia == "sábado" or dia == "domingo":
    print("Final de semana — descanse!")
else:
    print("Dia de trabalho")
````

````python
# not — inverte a condição
logado = False

if not logado:
    print("Faça login para continuar")
````

## 6. Condicionais aninhadas

Você pode colocar um `if` dentro de outro. Use com moderação — mais de 2 níveis costuma indicar que o código pode ser simplificado.

````python
# Verificação de acesso a um sistema
usuario_ativo = True
senha_correta = True
tem_permissao = False

if usuario_ativo:
    if senha_correta:
        if tem_permissao:
            print("Acesso total")
        else:
            print("Login ok, mas sem permissão para esta área")
    else:
        print("Senha incorreta")
else:
    print("Usuário inativo")
````

## 7. Operador ternário — `if` em uma linha

Para casos simples, Python permite escrever o if/else em uma única linha:

```python
valor = resultado_se_true if condição else resultado_se_false
```

````python
# Versão tradicional
idade = 20
if idade >= 18:
    status = "maior de idade"
else:
    status = "menor de idade"
print(status)

# Versão com operador ternário (equivalente)
status = "maior de idade" if idade >= 18 else "menor de idade"
print(status)
````
