# Python Básico — Funções

## O que vamos aprender
Funções são **blocos de código reutilizáveis** — você escreve uma vez e usa quantas vezes quiser.

Analogia: uma função é como uma receita de bolo. Você escreve a receita uma vez e pode fazer o bolo quantas vezes quiser, com ingredientes diferentes.

## 1. Por que usar funções?

Sem funções, código repetido fica assim:

````python
# Sem função — código repetido
preco1 = 100
desconto1 = preco1 * 0.10
print(f"Preço: R${preco1} | Desconto: R${desconto1} | Final: R${preco1 - desconto1}")

preco2 = 250
desconto2 = preco2 * 0.10
print(f"Preço: R${preco2} | Desconto: R${desconto2} | Final: R${preco2 - desconto2}")

preco3 = 80
desconto3 = preco3 * 0.10
print(f"Preço: R${preco3} | Desconto: R${desconto3} | Final: R${preco3 - desconto3}")
````

````python
# Com função — escreve uma vez, usa quantas quiser
def calcular_desconto(preco):
    desconto = preco * 0.10
    print(f"Preço: R${preco} | Desconto: R${desconto} | Final: R${preco - desconto}")

calcular_desconto(100)
calcular_desconto(250)
calcular_desconto(80)
````

## 2. Definindo e chamando funções

Sintaxe:
```python
def nome_da_funcao():
    # corpo da função
```

- `def` — palavra que avisa o Python que vem uma função
- `nome_da_funcao` — nome em `snake_case`
- `()` — parênteses (podem conter parâmetros)
- `:` — dois pontos obrigatórios
- corpo — indentado com 4 espaços

````python
# Definindo uma função simples
def saudar():
    print("Olá! Bem-vindo ao Python.")

# Chamando a função (executando)
saudar()
saudar()  # pode chamar quantas vezes quiser
saudar()
````

## 3. Parâmetros e argumentos

**Parâmetros** são as variáveis que a função espera receber (definidas no `def`).
**Argumentos** são os valores que você passa quando chama a função.

Analogia: parâmetro é o espaço na receita ("ingrediente"), argumento é o que você coloca ("farinha").

````python
# Função com um parâmetro
def saudar(nome):          # 'nome' é o parâmetro
    print(f"Olá, {nome}!")

saudar("Paulo")            # "Paulo" é o argumento
saudar("Maria")
saudar("João")
````

````python
# Função com múltiplos parâmetros
def calcular_area_retangulo(largura, altura):
    area = largura * altura
    print(f"Área: {area} m²")

calcular_area_retangulo(5, 3)   # largura=5, altura=3
calcular_area_retangulo(10, 2)  # largura=10, altura=2
````

## 4. `return` — devolvendo um valor

Até agora as funções apenas imprimem. Com `return`, a função **devolve um valor** que pode ser usado em outros cálculos.

Analogia: a calculadora mostra o resultado (print) vs. a calculadora devolve o resultado para você usar em outra conta (return).

````python
# Sem return — apenas imprime, não pode reutilizar o resultado
def dobrar_sem_return(numero):
    print(numero * 2)

# Com return — devolve o valor para ser usado
def dobrar(numero):
    return numero * 2

# Usando o valor retornado
resultado = dobrar(5)
print(resultado)           # 10
print(dobrar(3) + 1)       # 7 — usa o retorno diretamente
print(dobrar(dobrar(2)))   # 8 — função dentro de função
````

````python
# Função que retorna múltiplos valores
def calcular_preco_final(preco, percentual_desconto):
    desconto = preco * (percentual_desconto / 100)
    preco_final = preco - desconto
    return preco_final, desconto  # retorna dois valores

# Desempacotando os dois valores retornados
final, desc = calcular_preco_final(200, 15)
print(f"Desconto: R${desc:.2f}")
print(f"Preço final: R${final:.2f}")
````

## 5. Parâmetros com valor padrão

Você pode definir um valor padrão para um parâmetro. Se o argumento não for passado, usa o padrão.

````python
# desconto_percentual tem valor padrão de 10
def aplicar_desconto(preco, desconto_percentual=10):
    desconto = preco * (desconto_percentual / 100)
    return preco - desconto

print(aplicar_desconto(100))      # usa desconto padrão de 10% → 90.0
print(aplicar_desconto(100, 20))  # usa desconto de 20% → 80.0
print(aplicar_desconto(100, 5))   # usa desconto de 5% → 95.0
````

## 6. Argumentos nomeados

Você pode passar argumentos pelo nome — isso melhora a legibilidade e permite passar fora de ordem.

````python
def criar_usuario(nome, email, admin=False):
    print(f"Usuário: {nome} | Email: {email} | Admin: {admin}")

# Passando por posição (ordem importa)
criar_usuario("Paulo", "paulo@email.com")

# Passando por nome (ordem não importa)
criar_usuario(email="maria@email.com", nome="Maria", admin=True)

# Misturando — posicionais primeiro, depois nomeados
criar_usuario("João", "joao@email.com", admin=True)
````

## 7. Escopo de variáveis

**Escopo local**: variável criada dentro de uma função — só existe dentro dela.
**Escopo global**: variável criada fora de funções — existe em todo o arquivo.

Regra geral: prefira passar valores como parâmetros e retornar resultados, em vez de usar variáveis globais.

````python
mensagem_global = "Sou global"  # variável global

def minha_funcao():
    mensagem_local = "Sou local"   # variável local — só existe aqui
    print(mensagem_local)
    print(mensagem_global)         # pode ler variável global

minha_funcao()
print(mensagem_global)  # funciona

# print(mensagem_local)  # ← daria erro: variável local não existe fora da função
````

## 8. Docstrings — documentando funções

Uma boa função tem uma descrição de o que faz, o que recebe e o que retorna. Isso é feito com **docstring** — uma string entre `"""` logo após o `def`.

````python
def calcular_imc(peso, altura):
    """
    Calcula o Índice de Massa Corporal (IMC).

    Parâmetros:
        peso (float): peso em quilogramas
        altura (float): altura em metros

    Retorna:
        float: valor do IMC arredondado para 2 casas decimais
    """
    imc = peso / (altura ** 2)
    return round(imc, 2)

meu_imc = calcular_imc(75, 1.75)
print(f"IMC: {meu_imc}")

# help() exibe a docstring
help(calcular_imc)
````
