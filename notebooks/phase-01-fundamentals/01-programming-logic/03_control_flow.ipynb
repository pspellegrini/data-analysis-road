## Operadores de comparação

Antes de aprender decisões, precisamos saber como **comparar** valores.
O resultado de uma comparação é sempre `True` ou `False` (um `bool`).

| Operador | Significado | Exemplo | Resultado |
|---|---|---|---|
| `==` | igual a | `5 == 5` | `True` |
| `!=` | diferente de | `5 != 3` | `True` |
| `>` | maior que | `10 > 7` | `True` |
| `<` | menor que | `10 < 7` | `False` |
| `>=` | maior ou igual | `5 >= 5` | `True` |
| `<=` | menor ou igual | `3 <= 2` | `False` |

> **Atenção:** `=` é atribuição (`age = 25`). `==` é comparação (`age == 25`). São coisas diferentes!

````python
# testando operadores de comparação
temperature = 38.5

print(temperature > 37.5)   # True — está com febre?
print(temperature == 36.5)  # False — temperatura normal?
print(temperature <= 40.0)  # True — abaixo do limite crítico?
````

## Operadores lógicos

Para combinar duas ou mais comparações:

| Operador | Significado | Resultado |
|---|---|---|
| `and` | as duas condições precisam ser True | `True and False` → `False` |
| `or` | basta uma condição ser True | `True or False` → `True` |
| `not` | inverte o valor | `not True` → `False` |

````python
# verificando se um usuário pode acessar um serviço
is_logged_in = True
has_active_plan = False

# AND: as duas precisam ser True
print(is_logged_in and has_active_plan)   # False — não tem plano ativo

# OR: basta uma ser True
print(is_logged_in or has_active_plan)    # True — ao menos está logado

# NOT: inverte
print(not has_active_plan)                # True — confirma que NÃO tem plano
````

## Condicionais: `if`, `elif`, `else`

A estrutura condicional permite que o programa tome **caminhos diferentes** dependendo de uma condição.

```
SE (condição for verdadeira):
    faça isso
SENÃO SE (outra condição for verdadeira):
    faça aquilo
SENÃO:
    faça outra coisa
```

Em Python:
```python
if condição:
    # código executado se a condição for True
elif outra_condição:
    # código executado se a segunda condição for True
else:
    # código executado se nenhuma condição for True
```

> **Indentação é obrigatória!** O Python usa 4 espaços para definir o que está "dentro" do `if`.
> Se a indentação estiver errada, o programa vai falhar.

````python
# verificando se um passageiro pode embarcar
passenger_age = 16

if passenger_age >= 18:
    print("Embarque liberado.")  # só executa se a condição for True
else:
    print("Embarque não permitido — menor de idade.")  # executa se for False
````

````python
# classificando a temperatura com elif
temperature = 38.5

if temperature < 36.0:
    print("Hipotermia — temperatura abaixo do normal.")
elif temperature <= 37.5:
    print("Temperatura normal.")
elif temperature <= 39.0:
    print("Febre moderada — fique de olho.")
else:
    print("Febre alta — procure um médico.")

# o Python testa as condições de cima pra baixo e para na primeira que for True
````

````python
# combinando input() com if/else — programa interativo
grade = float(input("Digite sua nota (0 a 10): "))

if grade >= 7.0:
    result = "Aprovado"
elif grade >= 5.0:
    result = "Recuperação"
else:
    result = "Reprovado"

print(f"Nota: {grade:.1f} → {result}")
````

## Loop `for` — repetir para cada item

O `for` repete um bloco de código **uma vez para cada item** de uma sequência.

Pense numa lista de convidados: para cada convidado, envie um convite.

```python
for variável in sequência:
    # código executado a cada repetição
```

````python
# percorrendo uma lista de nomes
guests = ["Ana", "Bruno", "Carla", "Diego"]

for guest in guests:
    print(f"Convite enviado para {guest}.")
````

````python
# range() gera uma sequência de números
# range(5) → 0, 1, 2, 3, 4  (começa em 0, não inclui o 5)
# range(1, 6) → 1, 2, 3, 4, 5
# range(1, 11, 2) → 1, 3, 5, 7, 9  (pula de 2 em 2)

print("Tabuada do 3:")
for number in range(1, 11):
    result = 3 * number
    print(f"3 x {number} = {result}")
````

````python
# for com if — combinando repetição e decisão
# verificando quais números de 1 a 20 são pares

print("Números pares de 1 a 20:")
for number in range(1, 21):
    if number % 2 == 0:  # % retorna o resto da divisão — se for 0, é par
        print(number)
````

## Loop `while` — repetir enquanto uma condição for verdadeira

O `while` repete enquanto uma condição for `True`.
Use quando você **não sabe com antecedência** quantas repetições serão necessárias.

```python
while condição:
    # código executado enquanto a condição for True
```

> **Cuidado com loop infinito:** se a condição nunca se tornar `False`, o programa nunca para.
> Sempre garanta que algo dentro do `while` vai eventualmente encerrar o loop.

````python
# contagem regressiva — não sabemos de antemão quantas vezes vai rodar
countdown = 5

while countdown > 0:
    print(f"{countdown}...")
    countdown = countdown - 1  # reduz o contador a cada repetição

print("Lançamento!")
````

````python
# pedindo input até o usuário digitar algo válido
# caso de uso clássico do while: validação de entrada
password = ""

while password != "python123":
    password = input("Digite a senha: ")
    if password != "python123":
        print("Senha incorreta. Tente novamente.")

print("Acesso liberado!")
````

## `break` e `continue` — controlando loops

| Comando | O que faz |
|---|---|
| `break` | Interrompe o loop imediatamente |
| `continue` | Pula para a próxima iteração, ignorando o resto do bloco |

````python
# break — para de procurar assim que encontrar o primeiro produto com desconto
products = [
    {"name": "Notebook", "discount": False},
    {"name": "Mouse", "discount": False},
    {"name": "Teclado", "discount": True},
    {"name": "Monitor", "discount": True},
]

for product in products:
    if product["discount"]:
        print(f"Primeiro produto com desconto: {product['name']}")
        break  # encerra o loop — não precisamos checar o resto
````

````python
# continue — pula números negativos e processa só os positivos
values = [10, -3, 25, -8, 0, 42]

print("Processando apenas valores positivos:")
for value in values:
    if value <= 0:
        continue  # pula para o próximo item sem executar o print abaixo
    print(f"Valor: {value}")
````
