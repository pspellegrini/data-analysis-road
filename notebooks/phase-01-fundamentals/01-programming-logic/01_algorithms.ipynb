Antes de escrever qualquer linha de código, precisamos entender **como pensar como um programador**.
Isso começa com um conceito que está em tudo: o algoritmo.

## O que é um programa?

Um **programa** é um conjunto de instruções que diz ao computador o que fazer, passo a passo.

Pense numa receita de bolo:

1. Pré-aqueça o forno a 180°C
2. Misture os ingredientes secos
3. Adicione os ovos e o leite
4. Despeje na forma
5. Leve ao forno por 35 minutos

O forno não improvisa — ele só faz o que você mandou. O computador também.
**A diferença é que o computador segue as instruções na velocidade da luz.**

## O que é um algoritmo?

Um **algoritmo** é uma sequência de passos finita, clara e ordenada para resolver um problema.

Você usa algoritmos o tempo todo sem perceber:

- O GPS calcula a rota mais curta até o seu destino → algoritmo
- O aplicativo do banco decide se aprova seu cartão → algoritmo
- O YouTube escolhe o próximo vídeo para você → algoritmo

### Três características de um bom algoritmo

| Característica | Significado | Exemplo de problema se faltar |
|---|---|---|
| **Finito** | Tem começo e fim definidos | Um loop que nunca para |
| **Claro** | Cada passo é inequívoco | "Adicione sal a gosto" — quanto é "a gosto"? |
| **Correto** | Resolve o problema proposto | Um GPS que te leva ao lugar errado |

## Exemplo: algoritmo para fazer um café

```
INÍCIO
  1. Pegar a cafeteira
  2. Adicionar 2 colheres de café no filtro
  3. Colocar 200ml de água no reservatório
  4. Ligar a cafeteira
  5. Aguardar até o indicador luminoso apagar
  6. Servir o café na xícara
FIM
```

Perceba:
- Cada passo é **específico** — não "coloque água", mas "200ml"
- A ordem **importa** — não dá pra servir antes de ligar
- Tem um **fim claro** — o café está na xícara

## Primeiro código Python: `print()`

A função `print()` exibe uma mensagem na tela.

**Sintaxe:** `print("mensagem")` — o texto sempre vai entre aspas.

Execute a célula abaixo clicando nela e pressionando **Shift + Enter**.

````python
# meu primeiro programa em Python
# o símbolo # inicia um comentário — o Python ignora essa linha
print("Olá, mundo!")
````

````python
# várias instruções em sequência já formam um algoritmo
# o Python executa de cima para baixo, uma linha por vez
print("Passo 1: pegar a cafeteira")
print("Passo 2: adicionar o café no filtro")
print("Passo 3: adicionar 200ml de água")
print("Passo 4: ligar a cafeteira")
print("Passo 5: aguardar")
print("Café pronto!")
````

## Comentários no código

O símbolo `#` marca um **comentário** — uma nota para humanos que o Python ignora completamente.

Bons comentários explicam o **porquê**, não o quê (o código já mostra o quê):

```python
# ruim — apenas repete o que o código já diz
print("aguardando")  # imprime 'aguardando'

# bom — explica a razão da decisão
print("aguardando")  # mínimo de 4 minutos para o café não ficar amargo
```

> **Convenção deste projeto:** comentários sempre em **português (pt-BR)**,
> nomes de variáveis e funções sempre em **inglês**.

````python
# você pode comentar uma linha para 'desativá-la' sem apagar
# isso é útil quando você está testando o código

print("essa linha executa normalmente")
# print("essa linha está desativada — o Python não a vê")
print("essa linha também executa")
````
