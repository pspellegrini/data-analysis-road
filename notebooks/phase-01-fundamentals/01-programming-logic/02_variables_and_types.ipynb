## O que é uma variável?

Uma **variável** é um espaço na memória do computador onde você guarda um valor e dá um nome a ele.

Pense numa caixa com uma etiqueta:
- A **etiqueta** é o nome da variável (`age`, `name`, `price`)
- O **conteúdo da caixa** é o valor (`25`, `"Ana"`, `99.90`)
- Você pode trocar o conteúdo da caixa a qualquer momento

O sinal `=` não significa "igual" em matemática — significa **"armazene este valor nesta variável"**.

````python
# criando variáveis — o Python cria a 'caixa' e coloca o valor dentro
name = "Ana"
age = 25
height = 1.68

# exibindo os valores com print()
print(name)
print(age)
print(height)
````

````python
# variáveis podem ser atualizadas — o conteúdo da caixa muda
score = 0
print(score)  # exibe 0

score = 10
print(score)  # exibe 10 — o valor anterior foi substituído

score = score + 5  # pega o valor atual (10) e soma 5
print(score)  # exibe 15
````

## Regras para nomear variáveis

O Python tem regras obrigatórias e convenções de boas práticas:

| Tipo | Regra | Válido | Inválido |
|---|---|---|---|
| **Obrigatório** | Sem espaços | `total_price` | `total price` |
| **Obrigatório** | Não começa com número | `item1` | `1item` |
| **Obrigatório** | Só letras, números e `_` | `user_name` | `user-name` |
| **Convenção** | `snake_case` (minúsculas + `_`) | `first_name` | `FirstName` |
| **Convenção** | Nome em inglês | `total_price` | `preco_total` |
| **Convenção** | Nome descritivo | `user_age` | `x`, `a1` |

````python
# nomes bons — descritivos e em snake_case
product_name = "Notebook"
unit_price = 3500.00
quantity_in_stock = 12

# nomes ruins — evite esses padrões em código real
x = "Notebook"    # o que é x? impossível saber sem contexto
preco = 3500.00   # em português (foge da convenção do projeto)
a1 = 12           # sem significado

print(product_name, unit_price, quantity_in_stock)
````

## Tipos de dados

Cada variável em Python tem um **tipo** — que define que classe de informação ela armazena
e o que você pode fazer com ela.

Os quatro tipos básicos:

| Tipo | Nome técnico | Exemplo | Quando usar |
|---|---|---|---|
| Número inteiro | `int` | `25`, `-3`, `1000` | Contagens, idades, quantidades |
| Número decimal | `float` | `1.68`, `99.90`, `-0.5` | Medidas, preços, porcentagens |
| Texto | `str` | `"Ana"`, `"São Paulo"` | Nomes, frases, códigos |
| Verdadeiro/Falso | `bool` | `True`, `False` | Condições, flags, estado |

````python
# int — números inteiros, sem vírgula
user_age = 28
items_in_cart = 3
floor_number = -1  # pode ser negativo

print(user_age, items_in_cart, floor_number)
````

````python
# float — números com parte decimal
# atenção: Python usa ponto (.) como separador decimal, não vírgula
product_price = 49.90
user_height = 1.75
tax_rate = 0.12  # 12%

print(product_price, user_height, tax_rate)
````

````python
# str — texto, sempre entre aspas simples ou duplas
first_name = "Carlos"
city = 'São Paulo'       # aspas simples também funcionam
postal_code = "01310-100"  # número entre aspas vira texto — não dá pra somar

print(first_name, city, postal_code)
````

````python
# bool — apenas dois valores possíveis: True ou False (com maiúscula!)
is_logged_in = True
has_discount = False
is_adult = True

print(is_logged_in, has_discount, is_adult)
````

## A função `type()`

`type()` revela qual é o tipo de uma variável ou valor.
Muito útil quando você recebe dados de fontes externas e precisa saber com o que está lidando.

````python
# descobrindo o tipo de cada variável
print(type(42))         # <class 'int'>
print(type(3.14))       # <class 'float'>
print(type("olá"))      # <class 'str'>
print(type(True))       # <class 'bool'>
````

````python
# atenção: 42 e "42" são tipos diferentes!
number = 42
text = "42"

print(type(number))  # int — dá pra somar
print(type(text))    # str — não dá pra somar numericamente

print(number + 8)    # resultado: 50
print(text + "8")    # resultado: "428" — concatena como texto!
````

## Operadores aritméticos

Com `int` e `float` você pode fazer cálculos:

| Operador | Operação | Exemplo | Resultado |
|---|---|---|---|
| `+` | Adição | `10 + 3` | `13` |
| `-` | Subtração | `10 - 3` | `7` |
| `*` | Multiplicação | `10 * 3` | `30` |
| `/` | Divisão (sempre float) | `10 / 3` | `3.333...` |
| `//` | Divisão inteira | `10 // 3` | `3` |
| `%` | Resto da divisão | `10 % 3` | `1` |
| `**` | Potência | `2 ** 8` | `256` |

````python
# calculando o total de uma compra
unit_price = 49.90
quantity = 3
discount_rate = 0.10  # 10% de desconto

subtotal = unit_price * quantity
discount = subtotal * discount_rate
total = subtotal - discount

print(subtotal)  # 149.7
print(discount)  # 14.97
print(total)     # 134.73
````

````python
# divisão inteira e resto — úteis no dia a dia de dados
total_minutes = 137

hours = total_minutes // 60    # quantas horas completas? (2)
minutes = total_minutes % 60   # quantos minutos sobraram? (17)

print(hours)    # 2
print(minutes)  # 17
````

## f-strings — textos com variáveis dentro

Para montar uma frase misturando texto fixo e variáveis, use **f-string**:
coloque `f` antes das aspas e envolva a variável com `{}`.

É a forma mais legível e moderna de formatar texto em Python.

````python
first_name = "Carlos"
user_age = 28
city = "São Paulo"

# sem f-string — verboso e difícil de ler
print("Olá, " + first_name + "! Você tem " + str(user_age) + " anos.")

# com f-string — limpo e direto
print(f"Olá, {first_name}! Você tem {user_age} anos e mora em {city}.")
````

````python
# f-strings também aceitam expressões e formatação de números
unit_price = 49.90
quantity = 3
total = unit_price * quantity

# :.2f formata o número com exatamente 2 casas decimais
print(f"Total da compra: R$ {total:.2f}")
````

## A função `input()` — recebendo dados do usuário

`input()` pausa o programa e espera o usuário digitar algo.
**Atenção:** tudo que `input()` retorna é `str` — converta se precisar de número.

````python
# pedindo o nome do usuário e respondendo
user_name = input("Qual é o seu nome? ")
print(f"Olá, {user_name}! Bem-vindo ao Python.")
````

````python
# convertendo a entrada para número antes de calcular
# int() converte str para inteiro, float() para decimal
age_text = input("Qual é a sua idade? ")
age = int(age_text)  # converte de str para int

birth_year = 2025 - age
print(f"Você nasceu por volta de {birth_year}.")
````
